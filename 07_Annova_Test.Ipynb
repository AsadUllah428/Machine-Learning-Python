# ANOVA Test (Analysis of Variance)

The ANOVA (Analysis of Variance) test is a statistical method used to determine whether there are any statistically significant differences between the means of three or more independent (unrelated) groups. It is an extension of the t-test, which is limited to comparing only two groups.

## When to Use ANOVA?
- When you have three or more groups to compare.
- When your dependent variable is continuous (e.g., height, weight, score).
- When your independent variable is categorical (e.g., different treatments, brands, or groups).

## Types of ANOVA
- **One-way ANOVA:** Tests the effect of a single factor on a continuous outcome.
- **Two-way ANOVA:** Tests the effect of two factors and their interaction.

---

## Example: One-way ANOVA
Suppose you want to test whether three different diets have different effects on weight loss. You collect weight loss data from three groups, each following a different diet.

```python
import seaborn as sns
import scipy.stats as stats

# Load iris dataset
df1 = sns.load_dataset('iris')

# Extract groups
setosa = df1[df1['species'] == 'setosa']['sepal_length']
versicolor = df1[df1['species'] == 'versicolor']['sepal_length']
virginica = df1[df1['species'] == 'virginica']['sepal_length']

# One-way ANOVA
f_stat, p_val = stats.f_oneway(setosa, versicolor, virginica)


# Print the results
print("\nANOVA results (Setosa vs Versicolor vs Virginica):")
print(f"P-value = {p_val:.3f}")   # P-value < 0.05 suggests strong evidence against the null hypothesis

# Interpret the results
print("\nInterpretation:")
if p_val < 0.05:
    print("Since p-value < 0.05, we reject the null hypothesis.")
    print("There is a significant difference in sepal lengths among the three species.")
else:
    print("Since p-value >= 0.05, we fail to reject the null hypothesis.")
    print("There is not enough evidence to conclude a significant difference in sepal lengths among the species.")

```

```python
# Visualize the distributions to better understand the data
import matplotlib.pyplot as plt
import seaborn as sns

# --- Boxplot for all three species ---
plt.figure(figsize=(10, 6))
sns.boxplot(x='species', y='sepal_length', data=df1, palette="Set2")
plt.title('Sepal Length Distribution by Species')
plt.ylabel('Sepal Length (cm)')
plt.xlabel('Species')
plt.grid(True, axis='y', linestyle='--', alpha=0.7)
plt.show()

# --- Histogram for all three species ---
plt.figure(figsize=(12, 6))
for species in df1['species'].unique():
    subset = df1[df1['species'] == species]['sepal_length']
    plt.hist(subset, bins=15, alpha=0.5, label=species)

plt.title('Histogram of Sepal Lengths by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()

# --- Print Descriptive Statistics for each species ---
print("\nDescriptive Statistics:")
for species in df1['species'].unique():
    subset = df1[df1['species'] == species]['sepal_length']
    print(f"\n{species.capitalize()}:")
    print(f"Mean: {subset.mean():.2f}")
    print(f"Standard Deviation: {subset.std():.2f}")
    print(f"Sample Size: {len(subset)}")
```
